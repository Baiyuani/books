## return
```go
func FunctionName(param1 type, param2 type) type {
  // code to be executed
  return output
}
```

```go
package main
import ("fmt")

func myFunction(x int, y int) int {
  return x + y
}

func main() {
  fmt.Println(myFunction(1, 2))
}
```

#### 为返回值命名
```go
package main
import ("fmt")

func myFunction(x int, y int) (result int) {
  result = x + y
  return
}

func main() {
  fmt.Println(myFunction(1, 2))
}
```

```go
//也可以这样写
package main
import ("fmt")

func myFunction(x int, y int) (result int) {
  result = x + y
  return result
}

func main() {
  fmt.Println(myFunction(1, 2))
}
```

#### 将返回值存储在变量中
```go
package main
import ("fmt")

func myFunction(x, y int) (result int) {
  result = x + y
  return
}

func main() {
  total := myFunction(1, 2)
  fmt.Println(total)
}
```

#### 多个返回值
```go
package main
import ("fmt")

func myFunction(x int, y string) (result int, txt1 string) {
  result = x + x
  txt1 = y + " World!"
  return
}

func main() {
  a, b := myFunction(5, "Hello")
  fmt.Println(a, b)
}
//----------------------------------------------------------------
```
```output
10 Hello World!
```

```go
//如果不想使用某些返回值，我们可以添加一个下划线 (_)，以省略该值
package main
import ("fmt")

func myFunction(x int, y string) (result int, txt1 string) {
  result = x + x
  txt1 = y + " World!"
  return
}

func main() {
   _, b := myFunction(5, "Hello")
  fmt.Println(b)
}
```

## defer
```go
//defer 语句会将函数推迟到外层函数返回之后执行。

//推迟调用的函数其参数会立即求值，但直到外层函数返回前该函数都不会被调用。

package main

import "fmt"

func main() {
	defer fmt.Println("world")

	fmt.Println("hello")
}
//----------------------------------------------------------------
hello
world
```

## defer栈
```go
package main

import "fmt"
//推迟的函数调用会被压入一个栈中。当外层函数返回时，被推迟的函数会按照后进先出的顺序调用。
func main() {
	fmt.Println("counting")

	for i := 0; i < 10; i++ {
		defer fmt.Println(i)
	}

	fmt.Println("done")
}
//----------------------------------------------------------------
counting
done
9
8
7
6
5
4
3
2
1
0
```

## 递归
```go
package main
import ("fmt")

func testcount(x int) int {
  if x == 11 {
    return 0
  }
  fmt.Println(x)
  return testcount(x + 1)
}

func main(){
  testcount(1)
}
//----------------------------------------------------------------
```

```go
package main

import "fmt"

func fact(n int) int {
    if n == 0 {
        return 1
    }
    return n * fact(n-1)
}

func main() {
    fmt.Println(fact(7))

    var fib func(n int) int

    fib = func(n int) int {
        if n < 2 {
            return n
        }

        return fib(n-1) + fib(n-2)
    }

    fmt.Println(fib(7))
}
```

## 可变参数函数
```go
package main

import "fmt"
//这是一个将任意数量的整数作为参数的函数。在函数内部，nums 的类型等同于 []int。我们可以调用 len(nums)，用范围迭代它，等等
func sum(nums ...int) {
    fmt.Print(nums, " ")
    total := 0

    for _, num := range nums {
        total += num
    }
    fmt.Println(total)
}

func main() {

    sum(1, 2)
    sum(1, 2, 3)

    nums := []int{1, 2, 3, 4}
    sum(nums...)
}
```

## 闭包Closures
```go
package main

import "fmt"
//这个函数 intSeq 返回另一个函数，我们在 intSeq 的主体中匿名定义它。返回的函数关闭变量 i 以形成闭包
func intSeq() func() int {
    i := 0
    return func() int {
        i++
        return i
    }
}

func main() {
    //调用 intSeq，将结果（一个函数）赋值给 nextInt。这个函数值捕获它自己的 i 值，它会在我们每次调用 nextInt 时更新
    nextInt := intSeq()

    fmt.Println(nextInt())
    fmt.Println(nextInt())
    fmt.Println(nextInt())

    newInts := intSeq()
    fmt.Println(newInts())
}
//----------------------------------------------------------------
```
```output
1
2
3
1
```
